```python
import os
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

from feature_extraction import color_hist, edge_detect, utils
DATA_CATEGORIES = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']
METHODS = ['gaussian', 'mean', 'median']
data_dir = 'dataset-resized'
```

## Loading the RGB + edge detection histogram images into a df

```python
def extract_color_histogram(row):
    ch = color_hist.ColorHist(row['im_path'], 'outputs')
    ch.get_color_hist()
    ch.get_corresponding_edge()
    ch.get_edge_map()
    ch.get_lbp()
    ch.merge_hists()
    row['merged'] = ch.merged
    return row
```

```python
colorHist_df = pd.DataFrame(utils.create_image_paths(data_dir, load_images=True, vectorize_images=True))
colorHist_df = colorHist_df.apply(extract_color_histogram, axis=1)
```

```python
colorHist_df.head()
```

## Loading the Edges into a df

```python
edgeDetected_df = pd.DataFrame(utils.load_output_images(source_file_names=data_dir, load_images=True, vectorize_images=True))
    
```

## Merging datasets together

```python
colorHist_df['image_name'] = colorHist_df['im_path'].apply(lambda x: x.split('/')[-1])
edgeDetected_df['image_name'] = edgeDetected_df['im_path'].apply(lambda x: x.split('/')[-1])
```

```python
combined_full_df = pd.merge(left=colorHist_df,
                     right=edgeDetected_df,
                     left_on="image_name",
                     right_on="image_name",
                     suffixes=("_colorHist", "_edgeDetected")
                    ).drop(['class_colorHist'], axis=1).rename(columns={'class_edgeDetected':'class'})
def extract_individual(row):
    row['lbp_hist'] = row['merged'][0]
    row['edges_hist'] = row['merged'][1]
    row['red_hist'] = row['merged'][2]
    row['green_hist'] = row['merged'][3]
    row['blue_hist'] = row['merged'][4]
    return row

combined_full_df = combined_full_df.apply(extract_individual, axis=1)
```

```python
combined_full_df.head()
```

```python
combined_full_df.columns
```

#### Feature engineering

Since we can't train directly on a column that's a 4d array, we flatten the 4d array into features, so it's represented in a very wide matrix.

```python
# reshaped_data = np.reshape(combined_full_df['merged'].values, (combined_full_df['merged'].shape[0], 4*512))
# reshaped_data = np.reshape(combined_full_df['merged'].values, (combined_full_df.shape[0], -1))
reshaped_data = [np.asarray(v).flatten() for v in combined_full_df['merged']]
columns = ['feature_{}'.format(i) for i in range(len(reshaped_data[0]))]
df_new = pd.DataFrame(reshaped_data, columns=columns)

# Concatenate the new DataFrame with the original DataFrame
df_final = pd.concat([combined_full_df.drop('merged', axis=1), df_new], axis=1)

```

```python
df_final.head()
```

## Train Models below



```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_columns = [
#     "vectorized_R", # red band of raw image
#     "vectorized_G", # blue band of raw image
#     "vectorized_B", # green band of raw image
    'merged',
    'edges_hist',
    'red_hist',
    'green_hist',
    'blue_hist'
    # edge detection use mean
#     'vectorized_mean_R','vectorized_mean_G','vectorized_mean_B',
    # edge detecion using gaussian
#     'vectorized_gaussian_R','vectorized_gaussian_G','vectorized_gaussian_B',
    # edge detection using median
#     'vectorized_median_R','vectorized_median_G','vectorized_median_B'
]
X_columns = [f for f in df_final.columns[-2048:]]
Y_column = "class"
```

### Autoencode categorical Y variable

```python
X_train, X_test, y_train, y_test = train_test_split(df_final[X_columns],
                                                    df_final[Y_column].astype('category'),
                                                    test_size=0.2,
                                                    random_state=42)
label_encoder = LabelEncoder()
```

```python
display(y_train)
y_train = label_encoder.fit_transform(y_train)

display(y_train)
```

```python
# xgbc = XGBClassifier(objective='multi:softmax', enable_categorical=True)
xgbc = XGBClassifier(objective='multi:softmax')

print(xgbc)

xgbc.fit(X_train, y_train)

# - cross validataion
scores = cross_val_score(xgbc, X_train, y_train, cv=5)
print("Mean cross-validation score: %.2f" % scores.mean())

kfold = KFold(n_splits=10, shuffle=True)
kf_cv_scores = cross_val_score(xgbc, X_train, y_train, cv=kfold )
print("K-fold CV average score: %.2f" % kf_cv_scores.mean())

y_pred = xgbc.predict(X_test)
```

```python
y_true_numeric = label_encoder.fit_transform(y_test)
cm = confusion_matrix(y_true_numeric,y_pred)
print(cm)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
labels = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']
reversed_labels = labels.copy()
reversed_labels.reverse()
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=reversed_labels)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```
